**Three points in a plane are given, not all on the same straight line. How many lines can be drawn which are equidistant from these points?**

_____

- First, we recall that [the distance from a point to a line is equal to the shortest distance between the two](https://en.wikipedia.org/wiki/Distance_from_a_point_to_a_line)

- Since all three points lie on the same plane, we can visualize them on the regular Cartesian plane

- Let's consider a simple example where we have the following 3 points:
    1. $p_{1} = (0, 0)$ (i.e. it lies at the origin)
    2. $p_{2} = (1, 0)$
    3. $p_{3} = (1, 1)$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
plt.rcParams['figure.figsize'] = 10, 10
```

```python
p1 = (0, 0)
p2 = (1, 0)
p3 = (1, 1)
```

```python
plt.scatter(p1[0], p1[1], s=100, label='p1')
plt.scatter(p2[0], p2[1], s=100, label='p2')
plt.scatter(p3[0], p3[1], s=100, label='p3')
plt.legend()
plt.show()
```

- So for these three points, we need to think of the possible lines we can draw that will be equidistant between the three
    - Let's consider a few random examples of lines
        - Line 1: A horizontal line at $y=-1$
        - Line 2: A diagonal line with slope 1 with y intercept of 0.5
        - Line 3: A vertical line at $x=0.2$

```python
plt.scatter(p1[0], p1[1], s=100, label='p1')
plt.scatter(p2[0], p2[1], s=100, label='p2')
plt.scatter(p3[0], p3[1], s=100, label='p3')

plt.axhline(-1, label=r'Line 1: $y=-1$', ls = 'dashed', c = 'r')
x = np.arange(-0.5, 1.5, 0.01)
y = (1)*x + (0.5)
plt.plot(x, y, label=r'Line 2: $y = x - 0.5$', ls = 'dashed', c = 'c')
plt.axvline(0.2, label=r'Line 3: $x=0.2$', ls = 'dashed', c = 'y')

plt.legend()
plt.show()
```

- As we can clearly see from the plot:
    - Line 1 is closer to $p1$ and $p2$ than $p3$
    - Line 2 is closer to $p1$ and $p3$ than $p2$
    - Line 3 is closer to $p1$ than $p2$ or $p3$

- But what else do we notice from the plots?

*Line 1 (i.e. red dotted line)*
- For Line 1, we notice that if all three points are on the same side of the line
    - When all three points are on the same side, we can see that there will always be one point that isn't as close as the others
        - *Why?*
            - Because if they were all equally close, then you could draw a straight line between the points, which would violate the condition described in the problem statement
    - **Therefore, the three points cannot be on the same side of the line**
        - So what does this mean then?
            - The only other alternative is that the line has two points on one side, and one on the other
                - *Think about it*: there cannot be 0 dots on one side (since that would mean 3 on the other), and there can't be 3 on one side, so the only other two options are 1 and 2 or 2 and 1 (which are equivalent)

*Line 2 (i.e. blue dotted line)*

- It just so happens that the line we defined has the same slope as the line created by connecting $p_{1}$ and $p_{3}$ (i.e. the blue and green dots)
    - This means that line 2 must be equally close to the two points

- As we saw for line 1, we know that for a line to be equidistant from all three points, we need it to between them
    - Therefore, we could shift line 2 down such that it lies equidistant from the three points

*Line 3 (i.e. the yellow dotted line)*

- This line combines our findings from the other two lines
    - Since the line is parallel to the chord created by connecting $p_{2}$ and $p_{3}$, it is equidistant from the two points
    - It has two points on one side, and one point on the other
        - **Therefore we could shift the line slightly to the right to make it equidistant from all three points**

**Summary of findings**

- For any three points (that don't lie on the same line), we can create a line that is equidistant from all three by the following algorithm:
    1. Pick any two points and calculate the slope of the chord between them
    2. Find the midpoint between the other point (i.e. the one not selected in step 1) and **either** of two points
    3. Draw a line with slope calculated in step 1 that runs through the midpoint from step 2

- Now, since there are $\binom{3}{2} = 3$ ways to select the two points in step 1 (and $\binom{1}{1}$ ways to select the other point in step 2), **there are 3 possible ways to draw a line that is equidistant from the 3 points**
